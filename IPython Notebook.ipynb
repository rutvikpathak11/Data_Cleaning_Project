### Replace all the Inconsistent reference to entities with proper reference

```python
df = df.replace(['NA', 'Na', 'na', 'nA'],"North America")
df
```

```python
df = df.replace(['SA', 'Sa', 'sa', 'sA'],"South America")
df
```

```python
df = df.replace(['Tx', 'tx', 'TX','texas'],"Texas")
df
```

### Fix the NULL or empty continent code field by replacing with appropriate values.

```python
for row in range(len(df.index)):
    #print (df['continent_code'].iloc[row])
    continent_Code = df['continent_code'].iloc[row]
    if(continent_Code == '' or continent_Code == 'null'):
        #print ("hi")
        country = df['country_name'].iloc[row]
        for row1 in range(len(df.index)):
            newCountry = df['country_name'].iloc[row1]
            if(country == newCountry and df['continent_code'].iloc[row1] != ''):
               # print("here: ", df['continent_code'].iloc[row1])
                df['continent_code'].iloc[row] = df['continent_code'].iloc[row1]

df.head(10)
```

### This will group dataset by city/town and display the number of Landslides that has occurred in that place

```python
#Group by city/town 

df['Count'] = df.groupby('city/town')['city/town'].transform('count')
df
```

### As a last step, replace all left out empty values with NA which denotes Not Applicable

```python
#Replace all left out empty values with 'NA'
df = df.replace([''],"NA")
df
```

### Replace all the outliers with certain standard time.

```python
#Replace all Night with 9:00 pm and Morning with 9:00 am

df = df.replace(['Night', 'Late Night'],"9:00 PM")
df = df.replace(['Morning', 'Early Morning'],"9:00 AM")
df
```
